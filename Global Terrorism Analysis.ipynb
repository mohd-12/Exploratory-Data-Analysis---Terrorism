```python
df.dropna(axis = 1, how='all')
df.head()
```

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State' ,'region_txt':'Region'},inplace=False)
df.head()
```

```python
df.corr()
```

```python
plt.rcParams['figure.figsize']=15,11
sns.countplot(x=df['iyear'],palette='plasma')
plt.ylabel('Number of attacks',size=15)
plt.xticks(rotation=90)
plt.xlabel('Attack year',size=15)
plt.title('Increase in terrorism',size=25)
```

```python
plt.rcParams['figure.figsize']=15,11
sns.countplot(x=df['iyear'],hue='success',data=df,palette='plasma')
plt.ylabel('Number of attacks',size=20)
plt.xticks(rotation=90)
plt.xlabel('Attack year',size=20)
plt.title('Increase in terrorism',size=25)
```

```python
plt.rcParams['figure.figsize']=15,11
sns.barplot(x=df['country_txt'].value_counts()[:20].index,y=df['country'].value_counts()[:20].values,palette = 'plasma')
plt.ylabel('Number of attacks',size=20)
plt.xticks(rotation=90)
plt.xlabel('Country attacked',size=20)
plt.title('Terrorism in Country',size=25)
```
